```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys, os
sys.stdout.flush()
```

```python
def read_file(filename):
    response = []
    for i, line in enumerate(open(filename)):
        if i == 0:
            doc_sentence = ' '.join(line.strip().split(',')[1:])
            print "DOC SENTENCE " + str(i) + ":", doc_sentence
        response.append(line.strip().split(',')[0])
    print "SUMMARY:", ' '.join(response[1:])
```

```python
def plot_heatmap(filename):
    plt.figure()
    df = pd.read_csv(filename)
    df = df.set_index('model')
    del df.index.name
    sns.heatmap(df, xticklabels=True, yticklabels=True, square=True)
    plt.show()
    read_file(filename)
```

```python
def plot_doc(id):
    dir_path = 'csv_files/' + id + '/'
    filelist = {}
    for filename in os.listdir(dir_path):
        int_id = int(filename.split('.')[0])
        filelist[int_id] = filename
    for f_item in sorted(filelist.items(), key = lambda d:d[0]):
        filename = dir_path + f_item[1]
        plot_heatmap(filename)
```

```python
plot_doc('8004')
```
